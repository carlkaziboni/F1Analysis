```python
import requests
import json
import datetime
```

```python
# Get session details
country_name = "Monaco"
session_name = "Qualifying"
year = 2025
session = requests.get(f"https://api.openf1.org/v1/sessions?country_name={country_name}&session_name={session_name}&year={year}")
session_key = session.json()[0]['session_key']
session_key
```

```python
# Laps Data for Lewis Hamilton
Lewis_hamilton_number = 44
laps_lewis = requests.get(f"https://api.openf1.org/v1/laps?session_key={session_key}&driver_number={Lewis_hamilton_number}")
print(laps_lewis.status_code)
laps_lewis.json()
```

```python
# Laps data for Charles Leclerc
Charles_leclerc_number = 16
laps_charles = requests.get(f"https://api.openf1.org/v1/laps?session_key={session_key}&driver_number={Charles_leclerc_number}")
print(laps_charles.status_code)
laps_charles.json()
```

```python
# Get car data for Lewis Hamilton
car_data = requests.get(f"https://api.openf1.org/v1/car_data?session_key={session_key}&driver_number={Lewis_hamilton_number}&speed>=1")
print(car_data.status_code)
car_data.json()
```

```python
# Get car data for Charles Leclerc
car_data_charles = requests.get(f"https://api.openf1.org/v1/car_data?session_key={session_key}&driver_number={Charles_leclerc_number}&speed>=1")
print(car_data_charles.status_code)
car_data_charles.json()
```

```python
# Get start of lap 2 and beginning of lap 3
start_lap_2 = laps_lewis.json()[1]['date_start']
start_lap_3 = laps_lewis.json()[2]['date_start']
laps_lewis.json()[1]['lap_number']
```

```python
print(f"Start of Lap 2: {start_lap_2}")
print(f"Start of Lap 3: {start_lap_3}")
# Get time difference between start of lap 2 and start of lap 3
start_lap_2_datetime = datetime.datetime.fromisoformat(start_lap_2)
start_lap_3_datetime = datetime.datetime.fromisoformat(start_lap_3)
time_diff = start_lap_3_datetime - start_lap_2_datetime
print(f"Time difference between Lap 2 and Lap 3: {time_diff}")
time_diff.total_seconds()
```

```python
# Get location data for Lewis between start of lap 2 and start of lap 3 (end of lap 2)
location_data_lap_2 = requests.get(f"https://api.openf1.org/v1/location?session_key={session_key}&driver_number={Lewis_hamilton_number}&date>={start_lap_2}&date<{start_lap_3}")
print(f"Status code: {location_data_lap_2.status_code}")
location_data = location_data_lap_2.json()
print(f"Found {len(location_data)} location records for lap 2")
location_data
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract x and y coordinates from the location data
x_coords = [point['x'] for point in location_data]
y_coords = [point['y'] for point in location_data]

# Create the plot
plt.figure(figsize=(12, 8))
plt.plot(x_coords, y_coords, 'b-', linewidth=2, alpha=0.7, label='Lewis Hamilton Lap 2')
plt.scatter(x_coords[0], y_coords[0], color='green', s=100, marker='o', label='Start of Lap 2', zorder=5)
plt.scatter(x_coords[-1], y_coords[-1], color='red', s=100, marker='s', label='End of Lap 2', zorder=5)

plt.title('Monaco Grand Prix Circuit - Lewis Hamilton Lap 2 Path', fontsize=14)
plt.xlabel('X Coordinate', fontsize=12)
plt.ylabel('Y Coordinate', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()
plt.axis('equal')  # This ensures the track proportions are correct
plt.tight_layout()
plt.show()

print(f"Track visualization complete!")
print(f"Start position: ({x_coords[0]}, {y_coords[0]})")
print(f"End position: ({x_coords[-1]}, {y_coords[-1]})")
print(f"Total data points: {len(x_coords)}")
```

```python
# Get start of lap 2 and beginning of lap 3 for Charles Leclerc
start_lap_2_charles = laps_charles.json()[1]['date_start']
start_lap_3_charles = laps_charles.json()[2]['date_start']

print(f"Charles Leclerc - Start of Lap 2: {start_lap_2_charles}")
print(f"Charles Leclerc - Start of Lap 3: {start_lap_3_charles}")

# Get time difference for Charles' lap 2
start_lap_2_charles_dt = datetime.datetime.fromisoformat(start_lap_2_charles)
start_lap_3_charles_dt = datetime.datetime.fromisoformat(start_lap_3_charles)
time_diff_charles = start_lap_3_charles_dt - start_lap_2_charles_dt
print(f"Charles Leclerc - Time for Lap 2: {time_diff_charles}")
print(f"Charles Leclerc - Lap 2 duration: {time_diff_charles.total_seconds():.3f} seconds")
```

```python
# Get location data for Charles Leclerc between start of lap 2 and start of lap 3 (end of lap 2)
location_data_lap_2_charles = requests.get(f"https://api.openf1.org/v1/location?session_key={session_key}&driver_number={Charles_leclerc_number}&date>={start_lap_2_charles}&date<{start_lap_3_charles}")
print(f"Charles Leclerc - Status code: {location_data_lap_2_charles.status_code}")
location_data_charles = location_data_lap_2_charles.json()
print(f"Charles Leclerc - Found {len(location_data_charles)} location records for lap 2")

if len(location_data_charles) > 0:
    print(f"Charles Leclerc - First location point: {location_data_charles[0]}")
    print(f"Charles Leclerc - Last location point: {location_data_charles[-1]}")
else:
    print("No location data found for Charles Leclerc")
```

```python
# Plot Charles Leclerc's lap 2 path individually
x_coords_charles = [point['x'] for point in location_data_charles]
y_coords_charles = [point['y'] for point in location_data_charles]

plt.figure(figsize=(12, 8))
plt.plot(x_coords_charles, y_coords_charles, 'r-', linewidth=2, alpha=0.7, label='Charles Leclerc Lap 2')
plt.scatter(x_coords_charles[0], y_coords_charles[0], color='green', s=100, marker='o', label='Start of Lap 2', zorder=5)
plt.scatter(x_coords_charles[-1], y_coords_charles[-1], color='red', s=100, marker='s', label='End of Lap 2', zorder=5)

plt.title('Monaco Grand Prix Circuit - Charles Leclerc Lap 2 Path', fontsize=14)
plt.xlabel('X Coordinate', fontsize=12)
plt.ylabel('Y Coordinate', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()
plt.axis('equal')  # This ensures the track proportions are correct
plt.tight_layout()
plt.show()

print(f"Charles Leclerc Track visualization complete!")
print(f"Start position: ({x_coords_charles[0]}, {y_coords_charles[0]})")
print(f"End position: ({x_coords_charles[-1]}, {y_coords_charles[-1]})")
print(f"Total data points: {len(x_coords_charles)}")
```

```python
# Overlay comparison: Lewis Hamilton vs Charles Leclerc Monaco Lap 2
plt.figure(figsize=(14, 10))

# Plot both drivers' paths
plt.plot(x_coords, y_coords, 'b-', linewidth=2.5, alpha=0.8, label='Lewis Hamilton (#44) - Lap 2')
plt.plot(x_coords_charles, y_coords_charles, 'r-', linewidth=2.5, alpha=0.8, label='Charles Leclerc (#16) - Lap 2')

# Mark start and end points for both drivers
plt.scatter(x_coords[0], y_coords[0], color='darkblue', s=120, marker='o', label='Lewis Start', zorder=5)
plt.scatter(x_coords[-1], y_coords[-1], color='darkblue', s=120, marker='s', label='Lewis End', zorder=5)
plt.scatter(x_coords_charles[0], y_coords_charles[0], color='darkred', s=120, marker='o', label='Charles Start', zorder=5)
plt.scatter(x_coords_charles[-1], y_coords_charles[-1], color='darkred', s=120, marker='s', label='Charles End', zorder=5)

plt.title('Monaco Grand Prix Circuit - Lewis Hamilton vs Charles Leclerc\nLap 2 Path Comparison', fontsize=16, pad=20)
plt.xlabel('X Coordinate (meters)', fontsize=12)
plt.ylabel('Y Coordinate (meters)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(loc='best', fontsize=10)
plt.axis('equal')  # This ensures the track proportions are correct

# Add some styling
plt.gca().set_facecolor('#f8f9fa')
plt.tight_layout()
plt.show()

print(" Monaco Circuit Comparison Complete!")
print("="*50)
print(f" Lewis Hamilton (#44):")
print(f"    Data points: {len(x_coords)}")
print(f"    Start: ({x_coords[0]}, {y_coords[0]})")
print(f"    End: ({x_coords[-1]}, {y_coords[-1]})")
print(f" Charles Leclerc (#16):")
print(f"    Data points: {len(x_coords_charles)}")
print(f"    Start: ({x_coords_charles[0]}, {y_coords_charles[0]})")
print(f"    End: ({x_coords_charles[-1]}, {y_coords_charles[-1]})")
print("="*50)
```

```python
# Get car data for lap 2 for Lewis Hamilton
car_data_lewis_lap_2 = requests.get(f"https://api.openf1.org/v1/car_data?session_key={session_key}&driver_number={Lewis_hamilton_number}&date>={start_lap_2}&date<{start_lap_3}")
print(f"Status code for Lewis Hamilton car data: {car_data_lewis_lap_2.status_code}")
car_data_lewis_lap_2_json = car_data_lewis_lap_2.json()
print(f"Found {len(car_data_lewis_lap_2_json)} car data records for Lewis Hamilton Lap 2")
car_data_lewis_lap_2_json[:5]  # Display first 5 records
```

```python
# plot brake, throttle, speed, rpm and n_gear for Lewis Hamilton Lap 2

plt.figure(figsize=(14, 10))
plt.subplot(5, 1, 1)
plt.plot([record['date'] for record in car_data_lewis_lap_2_json], [record['brake'] for record in car_data_lewis_lap_2_json], label='Brake', color='blue')
plt.title('Lewis Hamilton Lap 2 - Brake')
plt.ylabel('Brake (%)')
plt.grid(True)

plt.subplot(5, 1, 2)
plt.plot([record['date'] for record in car_data_lewis_lap_2_json], [record['throttle'] for record in car_data_lewis_lap_2_json], label='Throttle', color='green')
plt.title('Lewis Hamilton Lap 2 - Throttle')
plt.ylabel('Throttle (%)')
plt.grid(True)

plt.subplot(5, 1, 3)
plt.plot([record['date'] for record in car_data_lewis_lap_2_json], [record['speed'] for record in car_data_lewis_lap_2_json], label='Speed', color='orange')
plt.title('Lewis Hamilton Lap 2 - Speed')
plt.ylabel('Speed (km/h)')
plt.grid(True)

plt.subplot(5, 1, 4)
plt.plot([record['date'] for record in car_data_lewis_lap_2_json], [record['rpm'] for record in car_data_lewis_lap_2_json], label='RPM', color='red')
plt.title('Lewis Hamilton Lap 2 - RPM')
plt.ylabel('RPM')
plt.grid(True)

plt.subplot(5, 1, 5)
plt.plot([record['date'] for record in car_data_lewis_lap_2_json], [record['n_gear'] for record in car_data_lewis_lap_2_json], label='Gear', color='purple')
plt.title('Lewis Hamilton Lap 2 - Gear')
plt.ylabel('Gear')
plt.xlabel('Time')
plt.grid(True)

plt.tight_layout()
plt.show()
```
